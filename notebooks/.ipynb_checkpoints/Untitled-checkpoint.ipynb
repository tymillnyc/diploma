```python
df5 = pd.read_csv('../datasets/foss/dataset_max_25.csv', index_col=0)
```

```python
df6 = pd.read_csv('../datasets/foss/merge_augment_parap_trnlt.csv', index_col=0)
```

```python
dict_count1 = dict(df6['category'].reset_index().groupby('category').count()['index'])
```

```python
dict_count = dict(df5['category'].reset_index().groupby('category').count()['index'])
```

```python
dict_count1
```

```python
dict_count
```

```python
before_augm  = dict_count.values()
after_augm = dict_count1.values()

after_augm1 = [x - y for x, y in zip(after_augm, before_augm)]
index = dict_count.keys()
df = pd.DataFrame({"before augm": before_augm,
                   "after augm": after_augm1}, index=index)
ax = df.plot.barh(stacked=True)
ax.legend(loc='center right', bbox_to_anchor=(1.32, 0.5))
```

```python
df.plot.barh(color={'before augm': "red", 'after augm': "green"})
```
